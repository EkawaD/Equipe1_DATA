```python
import pandas as pd #inclus matplotlib and numpy
```

```python
df = pd.read_csv('../input/young-people-survey/responses.csv')
```

```python
lonely5 = df[df['Loneliness'] == 5]
lonely4 = df[df['Loneliness'] == 4]
lonely3 = df[df['Loneliness'] == 3]
lonely2 = df[df['Loneliness'] == 2]
lonely1 = df[df['Loneliness'] == 1]
```

```python
df.iloc[:,0:1].mean()
```

Conclusion : On écoute plus de musique quand on se sent seul

```python
ln1_mean = lonely1.iloc[:,1:18].mean().sort_values(ascending=False)
ln2_mean = lonely2.iloc[:,1:18].mean().sort_values(ascending=False)
ln3_mean = lonely3.iloc[:,1:18].mean().sort_values(ascending=False)
ln4_mean = lonely4.iloc[:,1:18].mean().sort_values(ascending=False)
ln5_mean = lonely5.iloc[:,1:18].mean().sort_values(ascending=False)
lnall_mean = df.iloc[:,1:18].mean().sort_values(ascending=False)
```

```python
all_mean = pd.concat([ln1_mean,ln2_mean, ln3_mean, ln4_mean, ln5_mean, lnall_mean ], axis=1)
```

```python
tab = all_mean.rename(columns = {0:'Loneliness1', 1:'Loneliness2', 2:'Loneliness3', 3:'Loneliness4', 4:'Loneliness5', 5:'Average'})
```

```python
tab.plot.bar()
```
